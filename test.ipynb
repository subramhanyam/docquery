```python
import requests
import xml.etree.ElementTree as ET

def parse_sitemap(xml_url):
    response = requests.get(xml_url)
    response.raise_for_status()  # raise error if request fails

    # Parse XML
    root = ET.fromstring(response.content)

    # Find all <loc> tags
    links = [elem.text for elem in root.iter() if elem.tag.endswith('loc')]

    return links

# Example usage
sitemap_url = "https://quixy.com/post-sitemap1.xml"
all_links = parse_sitemap(sitemap_url)

# Print the links
for link in all_links:
    print(link)
```
